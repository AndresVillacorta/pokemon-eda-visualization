### Limpieza y preparación de la data

Retiramos columnas que no serán usadas.

```python
df = df.drop(
    [
        'abilities', 'base_egg_steps', 'base_happiness', 'japanese_name', 'percentage_male', 
        'pokedex_number','classfication','experience_growth','percentage_male', 'base_total', 'capture_rate'
    ], axis=1
)
```

```python
df.info()
```

Revisamos las columnas con valores faltantes.

```python
df.isnull().sum()
```

Completamos los valores faltantes puesto que no todos los pokemon son de 2 tipos diferentes.

```python
df['type2'] = df['type2'].fillna('None')
```

Analizamos las otras dos columnas con valores faltantes.

```python
df[df['weight_kg'].isnull()]
```

```python
df.info()
```

Creamos 3 columnas nuevas para conocer el acumulado de estadísticas de cada uno de los pokemon.

```python
# Estadísticas de ofensiva
df['punt_attack'] = df['attack'] + df['sp_attack'] + df['speed']
# Estadísitcas de defensa
df['punt_def'] = df['defense'] + df['sp_defense'] + df['hp']
# Total de estadísticas
df['stat'] = df['attack'] + df['sp_attack'] + df['speed'] + df['defense'] + df['sp_defense'] + df['hp']
```

Dividimos la data en 2 grupos: Pokemones Legendarios y No Legendarios ya que las estadísticas de los primeros tienden a ser más altas que los otros.

```python
normal = df[df['is_legendary'] == 0]
legend = df[df['is_legendary'] == 1]
```

### Visualización de data

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(15,3), dpi=300)
sns.countplot(data=df, x='type1')
plt.title('Cantidad de Pokemon por tipo-I', weight='bold')
plt.xlabel('Tipo de Pokemon',weight='bold')
plt.ylabel(' ')
plt.xticks(rotation=90);
```

Los tipos de pokemon que más se pueden encontrar son los de agua, normal y planta. Los tipos de pokemon que menos se pueden encontrar son los voladores, hada y hielo.

```python
plt.figure(figsize=(15,4), dpi=300)
sns.countplot(data=df, x='type2')
plt.title('Cantidad de Pokemon por tipo-II', weight='bold')
plt.xlabel('Tipo de Pokemon', weight='bold')
plt.ylabel(' ')
plt.xticks(rotation=90);
```

Se puede observar que poco más del 40% de los pokemon no cuentan con un segundo tipo. Por otro lado, es el tipo volador, tierra u veneno los que más se pueden encontrar.

**Top 3 No Legendarios por Estadística**

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15,8), dpi=300)

sns.barplot(data=normal.nlargest(3, 'attack'), x='name', y='attack', ax=axes[0,0])
axes[0,0].set_title('Top 3 ataque', weight='bold')
axes[0,0].set_xlabel('Nombre', weight='bold')
axes[0,0].set_ylabel('Ataque', weight='bold')
axes[0,0].set_ylim(160,190)

sns.barplot(data=normal.nlargest(3, 'defense'), x='name', y='defense', ax=axes[0,1])
axes[0,1].set_title('Top 3 defensa', weight='bold')
axes[0,1].set_xlabel('Nombre', weight='bold')
axes[0,1].set_ylabel('Defensa', weight='bold')
axes[0,1].set_ylim(150,240)

sns.barplot(data=normal.nlargest(3, 'sp_attack'), x='name', y='sp_attack', ax=axes[0,2])
axes[0,2].set_title('Top 3 ataque especial', weight='bold')
axes[0,2].set_xlabel('Nombre', weight='bold')
axes[0,2].set_ylabel('Ataque Especial', weight='bold')
axes[0,2].set_ylim(150,200)

sns.barplot(data=normal.nlargest(3, 'sp_defense'), x='name', y='sp_defense', ax=axes[1,0])
axes[1,0].set_title('Top 3 defensa especial', weight='bold')
axes[1,0].set_xlabel('Nombre', weight='bold')
axes[1,0].set_ylabel('Defensa Especial', weight='bold')
axes[1,0].set_ylim(140,240)

sns.barplot(data=normal.nlargest(3, 'speed'), x='name', y='speed', ax=axes[1,1])
axes[1,1].set_title('Top 3 velocidad', weight='bold')
axes[1,1].set_xlabel('Nombre', weight='bold')
axes[1,1].set_ylabel('Velocidad', weight='bold')
axes[1,1].set_ylim(130,170)

sns.barplot(data=normal.nlargest(3, 'hp'), x='name', y='hp', ax=axes[1,2])
axes[1,2].set_title('Top 3 puntos de vida', weight='bold')
axes[1,2].set_xlabel('Nombre', weight='bold')
axes[1,2].set_ylabel('Puntos de vida', weight='bold')
axes[1,2].set_ylim(160,270)

plt.subplots_adjust(wspace=0.3, hspace=0.3);
```

- **Ataque:** Heracross es el pokemon con más punto de ataque con una diferencia de 15 puntos con el segundo puesto.
- **Defensa:** Steelix, Schukle y Aggron comparten el podio como los pokemon con mayor defensa.
- **Ataque Especial:** Alakazam es el pokemon con mayor ataque especial de todos.
- **Defensa Especial:** Schukle es el pokemon con mayor defensa especial llevando una gran diferencia a los otros dos puestos.
- **Velocidad:** Ninjask ocupa el primero puesto como el pokemon más veloz de todos.
- **Puntos de vida:** Blissey y Chansey son los pokemon con más puntos de vida por una marcada diferencia con el tercer puesto.

Es importante recalcar que tanto Schukle como Alakazam aparecen en el top3 de 2 estadísticas.
